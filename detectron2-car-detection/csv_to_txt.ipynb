```python
import numpy as np
import pandas as pd
import os
```

#### vott csv (category=1) -> annotation/*.txt

```python
# csv 파일 읽어서 ndarrray로 저장 (image,xmin,ymin,xmax,ymax)
x=pd.read_csv('sha_car_project-export.csv',usecols=['image','xmin','ymin','xmax','ymax']).values

# ndarray를 'image'를 기준으로 정렬
for i,name in enumerate(x[:,0]):
    name=str(name).split('_')[1]
    num=int(name.split('.')[0])
    x[i][0]=num

x_df=pd.DataFrame(x) # 데이터 프레임으로 변경 후 정렬
x_df.columns=['num','xmin','ymin','xmax','ymax']
x_df=x_df.sort_values(by=['num'],axis=0)
x=x_df.to_numpy() # 다시 ndarray로 변경

# 디렉토리 생성 
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)

dir_name='anno_benz_car'
dir_path='./'+dir_name
createFolder(dir_path)

# 어노테이션 파일들 생성 
for x_one in x:
    path=dir_path+'/annotation_text_'+str(x_one[0])+'.txt'
    print(path,end=' -> ')
    with open(path,'a') as f:
        for idx2,val in enumerate(x_one):
            # 'num'은 파일 내용에서 제외 
            if idx2 != 0:
                # 마지막 원소(ymax)가 아닐 때
                if idx2 != len(x_one)-1:
                    f.write(str(val)+',')
                    print(val,end=',')
                # 마지막 원소(ymax)일 때
                else:
                    f.write(str(val)+'\n')
                    print(val,end='\n')
```
